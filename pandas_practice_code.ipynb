**Merging, joining, concatination**


```python
import numpy as np
import pandas as pd

dict1 =pd.DataFrame({
    "id": [101, 102, 103, 104, 105],
    "name": ["Anita Sharma", "Jayant Saha", "Akash Saha", "Riya Das", "Neha Gupta"],
    "age": [25, 30, None, 28, 26]
})

display(dict1)

dict2 =pd.DataFrame({
    "id": [101, 102, 103, 106, 107],
    "salary": [40000, 52000, None, 60000, 45000],
    "place": ["Delhi", "Kolkata", "Mumbai", "Chennai", None]
})

display(dict2)

#Merging
pd.merge(dict1,dict2, on ='id', how = 'inner')
pd.merge(dict1,dict2, on ='id', how = 'outer')
pd.merge(dict1,dict2, on ='id', how = 'right')
pd.merge(dict1,dict2, on ='id', how = 'left')

```

**Group By**

```python
import numpy as np
import pandas as pd

sales_data = {
    "Employee": ["Rahul", "Priya", "Rahul", "Amit", "Priya", "Amit", "Rahul"],
    "Department": ["IT", "HR", "IT", "Finance", "HR", "Finance", "IT"],
    "Month": ["Jan", "Jan", "Feb", "Feb", "Mar", "Mar", "Mar"],
    "Sales": [500, 300, 450, 600, 350, 650, 700]
}

df = pd.DataFrame(sales_data)
display(df)

group_by = df.groupby('Employee').sum('Sales')
display(group_by)

group_by_1 = df.groupby(['Employee', 'Department']).sum('Sales')
display(group_by_1)
```

```python
import numpy as np
import pandas as pd

df = pd.read_csv(r"/content/drive/MyDrive/Colab Notebooks/online_courses_feature_extraction.csv")
df

```

1 Create final_price after discount.

```python
df['final_price'] = df['course_price'] * (1 - df['discount_percent'] / 100)
display(df[['course_price', 'discount_percent', 'final_price']].head())
```

2 Create discount_amount

```python
df['discount_amount'] = df['course_price'] - df['final_price']
display(df[['course_price', 'discount_percent', 'final_price', 'discount_amount']].head())
```

Convert completed column:

Yes  1

No  0

```python
df['completed'] = df['completed'].map({'Yes': 1, 'No': 0})
display(df.head())
```

4 Extract:

signup_year

signup_month

signup_day

```python
df['signup_date'] = pd.to_datetime(df['signup_date'])
df['signup_year'] = df['signup_date'].dt.year
df['signup_month'] = df['signup_date'].dt.month
df['signup_day'] = df['signup_date'].dt.day
display(df[['signup_date', 'signup_year', 'signup_month', 'signup_day']].head())
```

Create is_active_user

1 if last_active_days  7

else 0

```python
df['is_active_user'] = df['last_active_days'].apply(lambda x: 1 if x <= 7 else 0)
display(df[['last_active_days', 'is_active_user']].head())
```

Create price_category

< 3500  Low

35005000  Medium

5000  High

```python
df['price_category'] = df['course_price'].apply(lambda x: 'Low' if x < 3500 else ('Medium' if x <= 5000 else 'High'))
display(df[['course_price', 'price_category']].head(10))
```

Create engagement_score


```python
df.head()
df['engagement_score'] = df['watch_time_minutes']/df['review_count']
display(df[['review_count','watch_time_minutes','engagement_score']])
```

8 Create high_rating_user

rating  4.5  1 else 0

```python
df.head()
df['high_rating_user'] = df['rating'].apply(lambda x: 1 if x >=  4.5 else 0)
display(df[['user_id', 'course_category','rating','high_rating_user']])
```

Create power_user
Condition:

completed == 1

rating  4.5

watch_time_minutes > 800

```python
df['power_user'] = ((df['completed'] == 1) & (df['rating'] >= 4.5) & (df['watch_time_minutes'] > 800)).astype(int)
display(df[['user_id', 'completed', 'rating', 'watch_time_minutes', 'power_user']])
```

Create inactive_risk

last_active_days > 10  1 else 0

```python
df['inactive_risk'] = (df['last_active_days'] > 10).astype(int)
display(df[['user_id', 'last_active_days', 'inactive_risk']].head(10))
```

Create price_per_minute

```python
df['price_per_minute'] = df['course_price'] / df['watch_time_minutes']
display(df[['course_price', 'watch_time_minutes', 'price_per_minute']].head())
```

12 Average rating per city
(Create new column with mapped average)

```python
df['city_avg_rating'] = df.groupby('city')['rating'].transform('mean')
display(df[['city', 'rating', 'city_avg_rating']].head(10))
```

Average completion rate per course_category

```python
completion_rate = df.groupby('course_category')['completed'].mean().reset_index()
completion_rate.columns = ['course_category', 'avg_completion_rate']
display(completion_rate)
```

Total revenue per course_category

```python
total_revenue = df.groupby('course_category')['final_price'].sum().reset_index()
total_revenue.columns = ['course_category', 'total_revenue']
display(total_revenue)
```

Most popular device (by count)

```python
device_counts = df['device'].value_counts().reset_index()
device_counts.columns = ['device', 'count']
display(device_counts)

most_popular = device_counts.iloc[0]['device']
print(f"The most popular device is: {most_popular}")
```

16 City-wise user count feature

```python
df['city_user_count'] = df.groupby('city')['user_id'].transform('count')
display(df[['city', 'user_id', 'city_user_count']].head(10))
```
