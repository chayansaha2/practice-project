Simulate rolling a die 500 times and count frequency of each number.

```python
import numpy as np
# Use 1 to 7 to include numbers 1 through 6
num = np.random.randint(1, 7, 500)
# Use np.unique to count frequencies
unique, counts = np.unique(num, return_counts=True)
result = dict(zip(unique, counts))
# Printing the dictionary will display it in a single line
print(f"Frequency: {result}")
```

Compute pairwise distances between rows of a 2D array.

```python
import numpy as np


arr = np.arange(1, 13).reshape(4, 3)
print("Original Array:")
print(arr)

diff = arr[:, np.newaxis, :] - arr[np.newaxis, :, :]


dist_matrix = np.sqrt(np.sum(diff**2, axis=-1))

print("\nPairwise Distance Matrix (Row i to Row j) using only NumPy:")
print(dist_matrix)
```

Remove rows from a 2D array that contain NaN values.

```python
import numpy as np

arr = np.array([
    [1, 2, 3],
    [4, np.nan, 6],
    [7, 8, 9]
])

clean_arr = arr[~np.isnan(arr).any(axis=1)]

print(clean_arr)
```
