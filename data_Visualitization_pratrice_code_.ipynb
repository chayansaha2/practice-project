**Plot**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


x = np.linspace(1,6,15)
print(x)
y = x ** 2
print(y)


plt.plot(x,y)
plt.title("Demo Practice Graph")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
```

**subplot**


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


x = np.linspace(3,8,10)
print(x)
y = x**3
print(y)

plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(y,x)

plt.subplot(2,2,3)
plt.plot(x**2,y)

plt.subplot(2,2,4)
plt.plot(x,y**3)





```

**object oriented way**


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x = np.linspace(3,5,15)
x
y = x**3
y

fig = plt.figure()
axes_1 = fig.add_axes([0.1,0.3,1,1])
axes_1.plot(x,y)
axes_2 = fig.add_axes([0.2,0.8,0.4,0.4])
axes_2.plot(x,y)
```

**seaborn**- histogram


```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('tips')
display(df)
plt.subplot(1,2,1)
sns.histplot(df['total_bill'], kde=True)
plt.subplot(1,2,2)
sns.histplot(df['tip'], kde=True)


```

**Catagorical plot** - countplot

**Matrixplot** - heatmap


```python
display(df)
tipscorr = df[['total_bill', 'tip', 'size']]
tipscorr.corr()
sns.heatmap(tipscorr.corr(),annot= True)
```

Pivot table heat map

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

flights = sns.load_dataset("flights")
display (flights)

pvt_flights = flights.pivot_table(values = 'passengers', index = 'month', columns= 'year')
display(pvt_flights)
sns.heatmap(pvt_flights)
```

**Regression Plot**
